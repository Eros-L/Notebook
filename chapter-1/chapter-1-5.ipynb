```python
# 重载算术运算符
class Vector2:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return "(%.2f, %.2f)" % (self.x, self.y)

    # +，加法重载
    def __add__(self, other):
        return Vector2(self.x + other.x, self.y + other.y)
    # -，减法重载
    def __sub__(self, other):
        return Vector2(self.x - other.x, self.y - other.y)
    # *，乘法重载
    def __mul__(self, other):
        return Vector2(self.x * other.x, self.y * other.y)
    # /，除法重载
    def __truediv__(self, other):
        return Vector2(self.x / other.x, self.y / other.y)
    # %，取模重载
    def __mod__(self, other):
        return Vector2(self.x % other.x, self.y % other.y)
    # **，幂重载
    def __pow__(self, power, modulo=None):
        return Vector2(self.x ** power, self.y ** power)
    # //，整除重载
    def __floordiv__(self, other):
        return Vector2(self.x // other.x, self.y // other.y)
    # @，点积重载
    def __matmul__(self, other):
        return self.x * other.y + self.y * other.y

    # +，反向加法重载
    def __radd__(self, other):
        return Vector2(other + self.x, other + self.y)
    # -，反向减法重载
    def __rsub__(self, other):
        return Vector2(other - self.x, other - self.y)
    # *，反向乘法重载
    def __rmul__(self, other):
        return Vector2(other * self.x, other * self.y)
    # /，反向除法重载
    def __rtruediv__(self, other):
        return Vector2(other / self.x, other / self.y)
    # %，反向取模重载
    def __rmod__(self, other):
        return Vector2(other % self.x, other % self.y)
    # **，反向幂重载
    def __rpow__(self, other):
        return Vector2(other ** self.x, other ** self.y)
    # //，反向整除重载
    def __rfloordiv__(self, other):
        return Vector2(other // self.x, other // self.y)
    # @，反向点积重载
    def __rmatmul__(self, other):
        return other * (self.x + self.y)

    # +，赋值加法重载
    def __iadd__(self, other):
        self.x += other.x
        self.y += other.y
        return self
    # -，赋值减法重载
    def __isub__(self, other):
        self.x -= other.x
        self.y -= other.y
        return self
    # *，赋值乘法重载
    def __imul__(self, other):
        self.x *= other.x
        self.y *= other.y
        return self
    # /，赋值除法重载
    def __itruediv__(self, other):
        self.x /= other.x
        self.y /= other.y
        return self
    # %，赋值取模重载
    def __imod__(self, other):
        self.x %= other.x
        self.y %= other.y
        return self
    # **，赋值幂重载
    def __ipow__(self, other):
        self.x **= other
        self.y **= other
        return self
    # //，赋值整除重载
    def __ifloordiv__(self, other):
        self.x //= other.x
        self.y //= other.y
        return self
    # @，赋值点积重载
    def __imatmul__(self, other):
        self.x = self.y = self.__matmul__(other)
        return self
    
v1, v2 = Vector2(12, 5), Vector2(8, 15)
print(v1, v2)

# 正向重载
print("正向重载")
print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
print(v1 / v2)
print(v1 % v2)
print(v1 ** 2)
print(v1 // v2)
print(v1 @ v2)

# 反向重载
# 当运算符的左侧为内建类型时，右侧为自定义类型进行算术匀算符运算时会出现TypeError错误
# 实现运算符重载,此时需要使用反向运算符的重载
print("反向重载")
print(2 + v1)
print(2 - v1)
print(2 * v1)
print(2 / v1)
print(2 % v1)
print(2 ** v1)
print(2 // v1)
print(2 @ v1)

# 赋值重载
print("赋值重载")
v1 += v2
print(v1)
v1 -= v2
print(v1)
v1 *= v2
print(v1)
v1 /= v2
print(v1)
v1 %= v2
print(v1)
v1 **= 2
print(v1)
v1 //= v2
print(v1)
v1 @= v2
print(v1)
```

```python
# 比较运算符
# 比较运算符的重载与算术运算符的重载类似
x, y, z = 2, 4, 8
# ==，等于
print(x == x)
print(x == y)
# !=，不等于
print(x != x)
print(x != y)
# >，大于
print(x > y)
print(z > y)
# <，小于
print(x < y)
print(z < y)
# >=，大于等于
print(x >= y)
# <=，小于等于
print(x <= y)
# 语法糖，连续比较
print(x == x <= x != y < z)
```

```python
# 位运算符
# 一般情况下不需要用到位运算符。但在某些算法题中，位运算符可以起到奇效。
# 其中一种应用是掩码（mask）
a = 0b00111100
b = 0b00001101
# &，按位与运算符
print('{:08b}'.format(a & b))
# |，按位或运算符
print('{:08b}'.format(a | b))
# ^，按位异或运算符
print('{:08b}'.format(a ^ b))
# ~，按位取反运算符
print('{:08b}'.format(~a))
# 由于 64 位机器上的整型有 64 个 bits
# 我们使用掩码来截取运算结果的前 8 位
mask = 2 ** 8 - 1   # 等价于 0b11111111
print('{:08b}'.format(~a & mask))
# <<，左移动运算符
print('{:08b}'.format(a << 2))
# >>，右移动运算符
print('{:08b}'.format(a >> 2))
```

```python
# 位运算符在算法题中的一个简单应用
# https://leetcode-cn.com/problems/single-number/
def singleNumber(nums: list):
    result = 0
    for num in nums:
        result ^= num
    return result
# 我们知道：除了某个元素只出现一次以外，其余每个元素均出现两次
# 因为 a ^ a == 0，我们可以设计出一个基于异或运算的算法
print(singleNumber([4,1,2,1,2]))
```

```python
# 逻辑运算符
a, b, c = 2, 4, 8
# 逻辑与
print(a < b and a < c)
print(a < b and c < b)
print(b < a and c < a)
# 逻辑或
print(a < b or a < c)
print(a < b or c < b)
print(b < a or c < a)
# 逻辑非
print(not (a < b))
print(not (c < b))
```

```python
# 成员运算符
a = 2
# 相等，is
# 需要注意 is 与 == 的区别：is 为地址比较；== 为值比较。此处因为 a 是 2 的引用，所以 is 与 == 的结果相同。
print(a is 2)
print(a == 2)
# 在处理 mutable 变量时，is 与 == 的效果不同。
l1 = l2 = [1, 2, 3]
print(l1 is [1, 2, 3])
print(l1 == [1, 2, 3])
print(l1 is l2)
print(l1 == l2)
# 不相等，is not
print(a is not 2)
```
